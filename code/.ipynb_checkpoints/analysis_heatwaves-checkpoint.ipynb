```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = cpv_filtered[['timespan']].values
y = cpv_filtered['HWMId_magnitude'].values

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.figure()
plt.scatter(cpv_filtered['timespan'], cpv_filtered['HWMId_magnitude'], alpha=0.3, label='Data')
plt.plot(cpv_filtered['timespan'], y_pred, color='red', label='Linear Fit')
plt.xlabel('Duration (days)')
plt.ylabel('Magnitude')
plt.title('Linear Regression: Duration vs Magnitude')
plt.legend()
plt.tight_layout()
# plt.savefig('results/analysis_final/duration_magnitude_regression.png')
plt.show()

print(f"Regression R^2: {model.score(X, y):.2f}")
```

```python
# Find events with duration or magnitude above the 99th percentile
long_outliers = cpv_filtered[cpv_filtered['timespan'] > cpv_filtered['timespan'].quantile(0.99)]
intense_outliers = cpv_filtered[cpv_filtered['HWMId_magnitude'] > cpv_filtered['HWMId_magnitude'].quantile(0.99)]

print("Long duration outliers:")
display(long_outliers[['time_amin', 'timespan', 'HWMId_magnitude']])

print("High magnitude outliers:")
display(intense_outliers[['time_amin', 'timespan', 'HWMId_magnitude']])
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model_poly = LinearRegression()
model_poly.fit(X_poly, y)
y_poly_pred = model_poly.predict(X_poly)

plt.figure()
plt.scatter(cpv_filtered['timespan'], cpv_filtered['HWMId_magnitude'], alpha=0.3, label='Data')
plt.plot(cpv_filtered['timespan'], y_poly_pred, color='green', label='Quadratic Fit')
plt.xlabel('Duration (days)')
plt.ylabel('Magnitude')
plt.title('Quadratic Regression: Duration vs Magnitude')
plt.legend()
plt.tight_layout()
# plt.savefig('results/analysis_final/duration_magnitude_quadratic.png')
plt.show()

print(f"Quadratic Regression R^2: {model_poly.score(X_poly, y):.2f}")
```
